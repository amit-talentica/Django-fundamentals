# Handling a request

```python
When Django server receives the request from a browser it takes several steps to handle the request

1.Request: "GET/HTTP/1.1"  (Django server receive request)
    
2.# in urls.py
url(r' ^$,welcome)          (Find a matching url mapping)

3.#in views.py
    def welcome(request):
    return HttpResponse("Hello World")   (Call corresponding view function)

Note: How URls are mapped??
    When urls request comes to django server.Django looks in urls.py then urls maps with regular expression,then maps urls with 
    view function.
    And then views function takes the request and generates response for that request
```

# Models and Migration

```python
Models are the python classes which are used to mapped with database tables
Migration code will be generated by Django but not always
command to check migrations: python manage.py showmigrations
To run migration: python manage.py migrate
```

```python
Django apps are python packages that contains models,views,urls,etc
These are reusable
```

# Migration Workflow

```python
1.change model code
2.Generate migration script
3.check migration script (python manage.py makemigrations)
4.python manage.py migrate
```
